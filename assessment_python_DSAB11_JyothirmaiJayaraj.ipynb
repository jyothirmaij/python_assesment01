Challenge 1 (10 Marks)
Employee Attendance Bonus Eligibility
Scenario

A company provides a monthly attendance bonus to employees who meet the following criteria:

Attendance percentage must be 90% or above.
Late arrivals must not exceed 2.

Write a function that determines whether an employee is eligible for the attendance bonus.


```python
def check_bonus(attendance, late_count):
    # TODO 1:
    # Return True if employee is eligible
    if attendance >= 90 and late_count <= 2:
        return True

    # TODO 2:
    # Otherwise return False
    else:
        return False

attendance = float(input("Attendance Percentage: "))
late_count = int(input("Late Arrivals: "))

# TODO 3:
# Call the function
check_bonus(attendance, late_count)
# TODO 4:
# Print
# Eligible for Bonus
# OR
# Not Eligible for Bonus
print("Eligible for Bonus" if check_bonus(attendance, late_count) else "Not Eligible for Bonus")


# Sample Input
# 94
# 1
# Output
# Eligible for Bonus
```

Challenge 2 (10 Marks)
Customer Feedback Analyzer
Scenario

An e-commerce company records customer ratings after each purchase.

The rating is converted into feedback labels.

Rating	Feedback
5	Excellent
4	Good
3	Average
1-2	Poor

The program should process multiple customer records and display the final feedback summary.



```python
def get_feedback(rating):
    # TODO 1
    # Return feedback
    if rating == 5:
        return "Excellent"
    elif rating == 4:
        return "Good"
    elif rating == 3:
        return "Average"
    elif rating <= 2:
        return "Poor"
    else:
        return "Invalid Rating" # Added to handle ratings outside 1-5

def feedback_summary():
    n = int(input("Enter number of customers: "))

    feedback = {}

    for i in range(n):
        customer = input('enter customer name:')
        rating = int(input('enter customer rating:'))

        # TODO 2
        # Find feedback by calling the get_feedback function
        customer_feedback = get_feedback(rating)

        # TODO 3
        # Store customer and feedback
        feedback[customer] = customer_feedback

    print("\nCustomer Feedback Summary")

    # TODO 4
    # Display all customer feedback
    for customer, fb in feedback.items(): # Used 'fb' to avoid shadowing outer 'feedback'
        print(customer + " : " + fb)


feedback_summary()


# Sample Input
# 3
# Alice
# 5
# John
# 3
# Maria
# 2
# Output
# Customer Feedback Summary

# Alice : Excellent
# John : Average
# Maria : Poor
```

Challenge 3 (20 Marks)
Sales Performance Dashboard
Scenario

A retail company wants to generate a quick sales summary.

Each transaction contains:

Salesperson SalesAmount

Example

Asha 15000
Rahul 22000
Asha 8000
John 12000
Rahul 6000

The program should

Calculate total company sales.
Calculate total sales for each salesperson.
Identify the top performer.
Display the sales summary.

```python
def sales_dashboard():
    n = int(input("Enter number of sales records: "))

    sales = {}

    for i in range(n):
        name, amount = input("Enter name and amount: ").split()
        amount = int(amount)

        # TODO 1
        # Update salesperson sales
        if name in sales:
            sales[name] += amount
        else:
            sales[name] = amount

    # TODO 2
    # Calculate total company sales
    total_sales = sum(sales.values())

    top_salesperson = ""
    highest_sale = 0

    # TODO 3
    # Identify top performer
    for name, sale in sales.items():
        if sale > highest_sale:
            highest_sale = sale
            top_salesperson = name

    print("Total Company Sales:", total_sales)
    print("Top Performer:", top_salesperson)

    print("/n Sales Summary")

    # TODO 4
    # Display salesperson-wise sales
    for name, sale in sales.items():
        print(name + " :", sale)


sales_dashboard()


# Sample Input
# 5
# Asha 15000
# Rahul 22000
# Asha 8000
# John 12000
# Rahul 6000
# Output
# Total Company Sales: 63000

# Top Performer: Rahul

# Sales Summary

# Asha : 23000
# Rahul : 28000
# John : 12000
```
